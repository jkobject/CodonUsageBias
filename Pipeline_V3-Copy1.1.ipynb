What we are looking for here is wether or not the entropy values and/or CUBF values are also correlated with the 3D distances. we are basically implementing the work done by Alen Diament https://www.nature.com/articles/ncomms6876#ref23 and adding CUB values instead of Codon Usage Frequency. the difference is that the codon usage frequency is computed amongst the 64 different codons and thus transmit also information about the distribution of amino acids and thus about the underlying proteins encoded and their functions. with the Codon Usage Bias we are just conveying the information of usage of synonimous codons and not about anything related to the protein function (even if CUB could influence certain aspects of the protein)

Thus here, the question is: are proteins with similar CUB also expressed in similar places. this is highly unlikely and the signal would be dim, but it would hint on an idea that the tRNA pool might not be even in the nucleus and that given this idea and the assumption that mRNA are transcribed in the vicinity of their translation, then variation of tRNA densities in the nuclei might be correlated with the CUB which would lead to the idea that gene positions and genome conformation is also related to variation of  tRNA concentration in the nucleus
